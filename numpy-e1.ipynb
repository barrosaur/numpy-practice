### Task
The dataset is a 2D numpy array of 10 students and 5 subjects, with random integer scores from 60 to 100.

#### Calculate useful stats:
Answer these questions:
<ol>
  <li>What is the average score per student?</li>
  <li>What is the average score per subject?</li>
  <li>What is the highest and lowest score in the whole class?</li>
  <li>Which student got the highest total score?</li>
</ol>

```python
# dataset
import numpy as np 

np.random.seed(0) # reproducible results
grades = np.random.randint(50, 100, size=(10, 5))
print(grades)
```

```python
# get average score per student
i = 0
for student in grades:
  ave_grade_per_student = np.average(student)
  i += 1
  print(f"Average of student {i}: ", ave_grade_per_student)
```

```python
# get average score per subject

col = 0
subj_count = 1
while col < 5:
  subj_scores = grades[:, col]
  col += 1

  print(f"Average score of subject {subj_count}: ", np.average(subj_scores))
  subj_count += 1
```

```python
# highest and lowest
print("Highest score: ", np.max(grades))
print("Lowest score: ", np.min(grades))
```

```python
# student highest total score

# np.argmax() returns index
# axis=0 - vertical
# axis=1 - horizontal

total_score = grades.sum(axis=1)
student = np.argmax(total_score)
highest_score = np.max(total_score)
print(f"Student {student} has the highest total score of {highest_score}")
```

### Bonus Tasks
<ul>
  <li>add 5 extra credit points to every grade but cap at 100</li>
  <li>sort students by their total score (highest to lowest)</li>
  <li>find all students who scored below 70 in any subject</li>
  <li>reshape the array to (5,10) -- what does that represent?</li>
</ul>

```python
# add 5 extra credit points to every grade stop at 100
extra_credit_grades = grades + 5

extra_credit_grades[extra_credit_grades > 100] = 100
print(extra_credit_grades)
```

```python
# sort students by their total score high -> low
total_score = grades.sum(axis=1)
sorted = np.argsort(total_score)[::-1]

for i in sorted:
  print(f"Student {i + 1}: {total_score[i]}")
```

```python
# finding all students who scored below 70

# Output: student 1
#         student 2
print("Students who scored below 70 in any subject: ")

for i, row in enumerate(grades):
  if np.any(row < 70):
    print(f"Student {i + 1}")
```

```python
# Reshape
reshaped_grades = grades.reshape(5, 10)

print(reshaped_grades)
```

The reshaped grades array means that the scores per subject are the rows and the columns are the scores of each student
