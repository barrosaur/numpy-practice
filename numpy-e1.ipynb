### Task
The dataset is a 2D numpy array of 10 students and 5 subjects, with random integer scores from 60 to 100.

#### Calculate useful stats:
Answer these questions:
<ol>
  <li>What is the average score per student?</li>
  <li>What is the average score per subject?</li>
  <li>What is the highest and lowest score in the whole class?</li>
  <li>Which student got the highest total score?</li>
</ol>

```python
# dataset
import numpy as np 

np.random.seed(0) # reproducible results
grades = np.random.randint(50, 100, size=(10, 5))
print(grades)
```

```python
# get average score per student
i = 0
for student in grades:
  ave_grade_per_student = np.average(student)
  i += 1
  print(f"Average of student {i}: ", ave_grade_per_student)
```

```python
# get average score per subject

col = 0
subj_count = 1
while col < 5:
  subj_scores = grades[:, col]
  col += 1

  print(f"Average score of subject {subj_count}: ", np.average(subj_scores))
  subj_count += 1
```

```python
# highest and lowest
print("Highest score: ", np.max(grades))
print("Lowest score: ", np.min(grades))
```

```python
# student highest total score

for rows in grades:
  total_scores = np.array([np.sum(rows)])

  print(type(total_scores))
```
